```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('udemy_courses.csv')
df=df.drop_duplicates()
print(df)
```

```python
#no. of paid and unpaid courses

b=sns.countplot(x='is_paid', data=df)
b.bar_label(b.containers[0])
plt.show()
#paid courses is more as compare to unpaid courses
    
```

```python
c=sns.countplot(x='subject', data=df)
c.bar_label(c.containers[0])
plt.show()
#We analyze the course of web development is more

```

```python
c=sns.countplot(x='level', data=df)
c.bar_label(c.containers[0])
plt.show()
#no. of levels of courses
```

```python
sns.barplot(x='is_paid', y='num_subscribers',data=df)
plt.show()
```

```python
sns.barplot(x='level', y='num_reviews',data=df)
plt.show()
```

```python
sns.barplot(x='subject', y='price',data=df)
plt.show()
#The price of web development is more as compare to other subjects
```

```python
sns.barplot(x='subject', y='content_duration',data=df)
plt.show() #the content_duration of web development course is more
```
